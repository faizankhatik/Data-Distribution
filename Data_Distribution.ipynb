```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
 fig, axs = plt.subplots(3, 2, figsize=(14, 12))
 fig.suptitle('Histogram with KDE of Various Distribution', fontsize=16)

# Normal Distribution (Gaussian Distribution)
data_normal = np.random.normal(loc=0, scale=1, size=1000)
sns.histplot(data_normal, kde=True, ax=axs[0, 0], color='blue',stat='density',bins=30)
axs[0, 0].set_title('Normal Distribution')
axs[0, 0].grid(True)

# Skewed Distribution
data_right_skewed = np.random.exponential(scale=1, size=1000)
data_left_skewed = -np.random.exponential(scale=1, size=1000)

# Right Skewed histogram
sns.histplot(data_right_skewed, kde=True, ax=axs[0, 1], color='red',stat='density',bins=30, label='Right skewed')
# Left skewed histogram
sns.histplot(data_left_skewed, kde=True, ax = axs[0,1], color='green',stat='density',bins=30, label='Left skewed')
axs[0, 1].legend()
axs[0, 1].set_title('Skewed Distribution')
axs[0, 1].grid(True)

# Uniform Distribution
data_uniform = np.random.uniform(low=0, high=1, size=1000)
sns.histplot(data_uniform, kde=True, ax=axs[1, 0], color='purple',stat='density',bins=30)
axs[1,0].set_title('Uniform Distribution')
axs[1,0].grid(True)

# Bimodel Distribution
data_bimodel = np.concatenate([np.random.normal(loc=-2, scale=1, size=500), np.random.normal(loc=2, scale=1, size=500)])
sns.histplot(data_bimodel, kde=True, ax=axs[1, 1], color='orange',stat='density',bins=30)
axs[1,1].set_title('Bimodel Distribution')
axs[1,1].grid(True)

# Multimodel Distribution
data_multimodel = np.concatenate([np.random.normal(loc=-6, scale=1.5, size=500),
                                  np.random.normal(loc=0, scale=1, size=500),
                                  np.random.normal(loc=6, scale=1.5, size=500)])
sns.histplot(data_multimodel, kde=True, ax=axs[2, 0], color='brown',stat='density',bins=30)
axs[2,0].set_title('Multimodel Distribution')
axs[2,0].grid(True)

# Hide empty subplot
axs[2,1].axis('off')

# Adjust layout
plt.tight_layout(rect=[0,0,1,0.96])
plt.show()
```
